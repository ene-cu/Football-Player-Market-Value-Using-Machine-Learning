```python
F_data_cleaned = pd.read_csv("E:\\FootBall_Player_Prediction\\scraper\\src\\Fifa_21\\merge_all_data\\F_21_22_23_cleaned.csv")
```

```python
F_data_cleaned.head()
```

```python
F_data_cleaned.info()
```

```python
#visualize the missing data
def plot_nas(df: pd.DataFrame):
    if df.isnull().sum().sum() != 0:
        na_df = (df.isnull().sum() / len(df)) * 100      
        na_df = na_df.drop(na_df[na_df == 0].index).sort_values(ascending=False)
        missing_data = pd.DataFrame({'Missing Ratio %': na_df})
        plt.figure(figsize=(5, 4))
        missing_data.plot(kind="barh")
        plt.title('Missing Data Percentage')
        plt.xlabel('Missing Ratio %')
        plt.ylabel('Features')
        plt.show()
    else:
        print('No NAs found')
plot_nas(F_data_cleaned)
```

```python
# Filling the missing values by mean value for "value_M/€" column
grouped_means = F_data_cleaned.groupby('short_name')['value_M/€'].transform('mean')
F_data_cleaned['value_M/€'] = F_data_cleaned['value_M/€'].fillna(grouped_means)
overall_mean = F_data_cleaned['value_M/€'].mean()
F_data_cleaned['value_M/€'] = F_data_cleaned['value_M/€'].fillna(overall_mean)


# Filling the missing values by mean value for "wage_K/€" column
grouped_means = F_data_cleaned.groupby('short_name')['wage_K/€'].transform('mean')
F_data_cleaned['wage_K/€'] = F_data_cleaned['wage_K/€'].fillna(grouped_means)
overall_mean = F_data_cleaned['wage_K/€'].mean()
F_data_cleaned['wage_K/€'] = F_data_cleaned['wage_K/€'].fillna(overall_mean)


# Filling the missing values by mean value for "release_clause_M/€" column
grouped_means = F_data_cleaned.groupby('short_name')['release_clause_M/€'].transform('mean')
F_data_cleaned['release_clause_M/€'] = F_data_cleaned['release_clause_M/€'].fillna(grouped_means)
overall_mean = F_data_cleaned['release_clause_M/€'].mean()
F_data_cleaned['release_clause_M/€'] = F_data_cleaned['release_clause_M/€'].fillna(overall_mean)


# Filling the missing values by mean value for "defending_marking" column
grouped_means = F_data_cleaned.groupby('short_name')['defending_marking'].transform('mean')
F_data_cleaned['defending_marking'] = F_data_cleaned['defending_marking'].fillna(grouped_means)
overall_mean = F_data_cleaned['defending_marking'].mean()
F_data_cleaned['defending_marking'] = F_data_cleaned['defending_marking'].fillna(overall_mean)


# Filling the missing values by mean value for "mentality_positioning" column
grouped_means = F_data_cleaned.groupby('short_name')['mentality_positioning'].transform('mean')
F_data_cleaned['mentality_positioning'] = F_data_cleaned['mentality_positioning'].fillna(grouped_means)
overall_mean = F_data_cleaned['mentality_positioning'].mean()
F_data_cleaned['mentality_positioning'] = F_data_cleaned['mentality_positioning'].fillna(overall_mean)
```

```python
#visualize the missing data
def plot_nas(df: pd.DataFrame):
    if df.isnull().sum().sum() != 0:
        na_df = (df.isnull().sum() / len(df)) * 100      
        na_df = na_df.drop(na_df[na_df == 0].index).sort_values(ascending=False)
        missing_data = pd.DataFrame({'Missing Ratio %': na_df})
        plt.figure(figsize=(5, 4))
        missing_data.plot(kind="barh")
        plt.title('Missing Data Percentage')
        plt.xlabel('Missing Ratio %')
        plt.ylabel('Features')
        plt.show()
    else:
        print('No NAs found')
plot_nas(F_data_cleaned)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor  
import xgboost as xg 
```

```python
x = F_21_22_23_data.drop(['short_name','value_M/€','release_clause_M/€'],axis = 1)
y = F_21_22_23_data['value_M/€']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.3,random_state = 30)
```

```python
#Linear Regression Implementation
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
print(lr.intercept_)
```

```python
print(lr.coef_)
```

```python
lr_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

```python
LR_r2 = r2_score(y_test,lr_pred)*100
```

```python
print('r2 score: ',(LR_r2))
```

```python
LR_mse =  mean_squared_error(y_test,lr_pred)
```

```python
print(LR_mse)
```

```python
#Random Forest Regressor Implementation
rf = RandomForestRegressor()
rf.fit(x_train,y_train)
```

```python
rf_pred = rf.predict(x_test)
```

```python
RF_r2 = r2_score(y_test,rf_pred)*100
```

```python
print('r2 score: ',(RF_r2))
```

```python
RF_mse = mean_squared_error(y_test,rf_pred)
```

```python
print(RF_mse)
```

```python
#Decision Tree Regressor Implementation
regressor = DecisionTreeRegressor()
```

```python
regressor.fit(x_train,y_train) 
```

```python
DT_pred = regressor.predict(x_test) 
```

```python
DT_r2 = r2_score(y_test,DT_pred)*100
```

```python
print('r2 score: ',(DT_r2))
```

```python
DT_mse = mean_squared_error(y_test,DT_pred)
```

```python
print(DT_mse)
```

```python
#XG Boost Regressor Implementation
xgb = xg.XGBRegressor() 
```

```python
xgb.fit(x_train,y_train) 
```

```python
XGB_pred = xgb.predict(x_test) 
```

```python
XGB_r2 = r2_score(y_test,XGB_pred)*100
```

```python
print('r2 score: ',(XGB_r2))
```

```python
XGB_mse = mean_squared_error(y_test,XGB_pred)
```

```python
print(XGB_mse)
```

```python
model_results_comparison = pd.DataFrame({'Model': ['Linear Regression','Random Forest Regressor','Decision Tree Regressor','XGB Regressor'],'R_2 Score':[LR_r2,RF_r2,DT_r2,XGB_r2]})
```

```python
model_results_comparison.sort_values(by = 'R_2 Score',ascending = False)
```
